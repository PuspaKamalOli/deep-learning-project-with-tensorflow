```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Input,Conv2D, Dense, Flatten,Dropout, GlobalMaxPooling2D,BatchNormalization,MaxPooling2D
from tensorflow.keras.models import Model
```

```python
cifar10=tf.keras.datasets.cifar10
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
x_train,x_test=x_train/255,x_test/255
print(x_train.shape)
y_train,y_test=y_train.flatten(),y_test.flatten()
```

```python
#no of classes

k=len(set(y_train))
print("no of classes:",k)
```

```python
i=Input(shape=x_train[0].shape)
x=Conv2D( 32,(3,3),activation='relu',padding='same')(i)
x=BatchNormalization()(x)
x=Conv2D( 32,(3,3),activation='relu',padding='same')(x)
x=BatchNormalization()(x)
x=MaxPooling2D((2,2))(x)

i=Input(shape=x_train[0].shape)
x=Conv2D( 64,(3,3),activation='relu',padding='same')(i)
x=BatchNormalization()(x)
x=Conv2D( 64,(3,3),activation='relu',padding='same')(x)
x=BatchNormalization()(x)
x=MaxPooling2D((2,2))(x)
i=Input(shape=x_train[0].shape)
x=Conv2D( 128,(3,3),activation='relu',padding='same')(i)
x=BatchNormalization()(x)
x=Conv2D( 128,(3,3),activation='relu',padding='same')(x)
x=BatchNormalization()(x)
x=MaxPooling2D((2,2))(x)

x=Flatten()(x)
x=Dropout(0.2)(x)
x=Dense(1024,activation='relu')(x)
x=Dropout(0.2)(x)
x=Dense(k,activation='softmax')(x)
```

```python
model=Model(i,x)
```

```python
from keras.engine.training import optimizer
#compile and fit
#note make sure you are using GPU for this
model.compile(optimizer="adam",
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
r=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=15)
```

```python
plt.plot(r.history['accuracy'],label="accuracy")
plt.plot(r.history['val_accuracy'],label='val_accuracy')
```
